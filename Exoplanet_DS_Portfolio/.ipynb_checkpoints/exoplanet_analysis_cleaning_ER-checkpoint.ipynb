## Using the elbow method to find the optimal number of clusters  - Based On Course Content

```python
X = df.iloc[:,:].values
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters = i, init = "k-means++", random_state = 42) #k-means++ avoids 
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

#k-means++ does a weighted selection of centroids to avoid an inital centroid choice bias
```

```python
plt.scatter(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training the K-Means model on the dataset - Based On Course Content

Let's guess that n = 3 clusters makes the most sense!

```python
kmeans = KMeans(n_clusters = 3, init='k-means++', random_state=42)
y_pred = kmeans.fit_predict(X) #fit_predict will return a vector for each data point which cluster it belongs to
```

```python
y_pred
```

```python
X.shape
```

###Evaluate Clustering - My Work Beyond Course Content

```python
#Add Cluster Info To The DataFrame
df['Cluster']=y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```

## Visualising the clusters - Course Content

```python
# kmeans.cluster_centers_ - gives the x,y coordinates of the centroids

sz = 75

plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], s = sz, c = 'red', label = 'Cluster 0')
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], s = sz, c = 'blue', label = 'Cluster 1')
plt.scatter(X[y_pred == 2, 0], X[y_pred == 2, 1], s = sz, c = 'green', label = 'Cluster 2')
#plt.scatter(X[y_pred == 3, 0], X[y_pred == 3, 1], s = sz, c = 'cyan', label = 'Cluster 3')
#plt.scatter(X[y_pred == 4, 0], X[y_pred == 4, 1], s = sz, c = 'magenta', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = sz*2, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Planet Radius [$R_{E}]$')
plt.ylabel('Planet Mass [$M_{E}]$')
plt.legend()
plt.show()
```

```python
plt.scatter(range(1,nn), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
